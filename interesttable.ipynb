## Introduction
This notebook is used for calculating the total value of a 1000 dollar investment compounding at 3.5% annually over 30 years.  The formula used to calculate the interest value is: 
$Interest =\ 1000(1.035^{t})$
where t is years.

```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
interest = [(x,round(1000*(1.035**x),2),"$"+str(round(1000*(1.035**x)))) for x in linspace(0,30,31)]
```

```python
table = pd.DataFrame(interest)
table.columns = ['years','value_float','value_str']
```

```python
table
```

```python
plt.plot(table.years,table.value_float)
plt.xlabel(table.years.name)
plt.ylabel(table.value_float.name)
plt.legend()
plt.show()
```
